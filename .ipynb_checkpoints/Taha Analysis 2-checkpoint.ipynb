```python
def preprocess_data(data):
    ''' Change the categorical columns to numeric '''
    data = data.dropna()
    data = data.drop_duplicates()
    cols = data.columns
    for column in data.columns:
        if data[column].dtype == object:
            data[column] = data[column].astype('category').cat.codes
    scaler = MinMaxScaler()
    data = scaler.fit_transform(data)
    data = pd.DataFrame(data)
    data.columns = cols
    return data
```

```python
def plot_silhouette(X):
    range_n_clusters = [2, 3, 4, 5, 6, 7, 8]
    avgs = []
    for n_clusters in range_n_clusters:
        # Create a subplot with 1 row and 2 columns
        #fig.set_size_inches(18, 7)

        # The 1st subplot is the silhouette plot
        # The silhouette coefficient can range from -1, 1 but in this example all
        # lie within [-0.1, 1]
        plt.xlim([-0.1, 1])
        # The (n_clusters+1)*10 is for inserting blank space between silhouette
        # plots of individual clusters, to demarcate them clearly.
        plt.ylim([0, len(X) + (n_clusters + 1) * 10])

        # Initialize the clusterer with n_clusters value and a random generator
        # seed of 10 for reproducibility.
        clusterer = KMeans(n_clusters=n_clusters, random_state=10)
        cluster_labels = clusterer.fit_predict(X)

        # The silhouette_score gives the average value for all the samples.
        # This gives a perspective into the density and separation of the formed
        # clusters
        silhouette_avg = silhouette_score(X, cluster_labels)
        print("For n_clusters =", n_clusters,
              "The average silhouette_score is :", silhouette_avg)
        avgs.append(silhouette_avg)
        # Compute the silhouette scores for each sample
        sample_silhouette_values = silhouette_samples(X, cluster_labels)

        y_lower = 10
        for i in range(n_clusters):
            # Aggregate the silhouette scores for samples belonging to
            # cluster i, and sort them
            ith_cluster_silhouette_values = \
                sample_silhouette_values[cluster_labels == i]

            ith_cluster_silhouette_values.sort()

            size_cluster_i = ith_cluster_silhouette_values.shape[0]
            y_upper = y_lower + size_cluster_i

            color = cm.nipy_spectral(float(i) / n_clusters)
            plt.fill_betweenx(np.arange(y_lower, y_upper),
                              0, ith_cluster_silhouette_values,
                              facecolor=color, edgecolor=color, alpha=0.7)

            # Label the silhouette plots with their cluster numbers at the middle
            plt.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

            # Compute the new y_lower for next plot
            y_lower = y_upper + 10  # 10 for the 0 samples

        plt.title("The silhouette plot for the various clusters.")
        plt.xlabel("The silhouette coefficient values")
        plt.ylabel("Cluster label")

        # The vertical line for average silhouette score of all the values
        plt.axvline(x=silhouette_avg, color="red", linestyle="--")
        plt.show()
    avg_df = [({'number of clusters': i+2,'silhouette score': a}) for i,a in enumerate(avgs)]
    avg_df = pd.DataFrame(avg_df)
    return avg_df
```

```python
import itertools
color_iter = itertools.cycle(['navy', 'c', 'cornflowerblue', 'gold',
                              'darkorange'])
def plot_results(X, Y_, means, covariances, index, title, labels):
    splot = plt.subplot(2, 1, 1 + index)
    for i, (mean, covar, color) in enumerate(zip(
            means, covariances, color_iter)):
        v, w = linalg.eigh(covar)
        v = 2. * np.sqrt(2.) * np.sqrt(v)
        u = w[0] / linalg.norm(w[0])
        # as the DP will not use every component it has access to
        # unless it needs it, we shouldn't plot the redundant
        # components.
        if not np.any(Y_ == i):
            continue
        plt.scatter(X[Y_ == i, 0], X[Y_ == i, 1], .8, color=color)

        # Plot an ellipse to show the Gaussian component
        angle = np.arctan(u[1] / u[0])
        angle = 180. * angle / np.pi  # convert to degrees
        ell = mpl.patches.Ellipse(mean, v[0], v[1], 180. + angle, color=color)
        ell.set_clip_box(splot.bbox)
        ell.set_alpha(0.5)
        splot.add_artist(ell)

#     plt.xlim(-9., 5.)
#     plt.ylim(-3., 6.)
    plt.xticks(())
    plt.yticks(())
    plt.title(title)
    
def plot_gaus(X, n_components, labels):
    gmm = mixture.GaussianMixture(n_components=n_components, covariance_type='full').fit(X)
    plot_results(X, gmm.predict(X), gmm.means_, gmm.covariances_, 0,
                 'Gaussian Mixture K=' + str(n_components), labels)

    # Fit a Dirichlet process Gaussian mixture using five components
    dpgmm = mixture.BayesianGaussianMixture(n_components=5,
                                            covariance_type='full').fit(X)
    plt.show()
    return gmm.bic(X)
```

```python
class Clusterer():
    def __init__(self, method='KMeans'):
        self.clustering_labels = None
        self.clustering_scores = {}
    
    def fit_data(self, data, method='KMeans', n_clusters=2, label_name=None, is_gaus=False, include_ARI= True):
        if method == 'KMeans':
            clf = KMeans(n_clusters=n_clusters, random_state=0)
        elif method == 'GMM':
            clf = mixture.GaussianMixture(n_components=n_clusters, covariance_type='full').fit(data)
        try:
            return self.get_lineup_scores(data,clf, label_name, is_gaus=False, include_ARI= True)
        except:
            labels = data[label_name]
            pred_labels = clf.predict(data.drop([label_name], axis=1))
            return homogeneity_score(labels, pred_labels)
    
    def plot_elbow(self, data, K_range=range(1,9)):
        Sum_of_squared_distances = []
        for k in K_range:
            km = KMeans(n_clusters=k)
            km = km.fit(data)
            Sum_of_squared_distances.append(km.inertia_)
        plt.plot(K_range, Sum_of_squared_distances, 'bx-')
        plt.xlabel('k')
        plt.ylabel('Sum_of_squared_distances')
        plt.title('Elbow Method For Optimal k')
        plt.show()
    
    def get_lineup_scores(self, data, clf, label_name, is_gaus=False, include_ARI=True):
        labels = data[label_name]
        unlabeled_data = data.drop([label_name], axis=1)
        clustering_res = clf.fit(unlabeled_data)
        self.clustering_labels = clf.labels_ if not is_gaus else clf.predict(unlabeled_data)
        self.clustering_scores['homogeneity'], self.clustering_scores['completeness'], self.clustering_scores['v_measure'] = homogeneity_completeness_v_measure(labels, self.clustering_labels)
        if include_ARI:
            self.clustering_scores['ARI'] = adjusted_rand_score(labels, self.clustering_labels)
    
    def plot_gaus(self, data, labels, K_range=range(2,9)):
        bic_scores = []
        for n in K_range:
            bic_scores.append(plot_gaus(data, n, labels))
        return bic_scores
```

```python
def get_clustering_score(data, clf, label_name, is_gaus=False, include_ARI= True):
    labels = data[label_name]
    unlabeled_data = data.drop([label_name], axis=1)
    clustering_res = clf.fit(unlabeled_data)
    if not is_gaus:
        clustering_labels = clf.labels_
    else:
        clustering_labels = clf.predict(unlabeled_data)
    if not label_name:
        return clustering_labels
    if not include_ARI:
        return homogeneity_completeness_v_measure(labels, clustering_labels), clustering_labels, 
    else:
        return adjusted_rand_score(labels, clustering_labels), homogeneity_completeness_v_measure(labels, clustering_labels), clutering_labels
```

# Dimensionality Reduction

### PCA

```python
def appply_dim_reduction(data, clf, label_name, isPCA=False, isICA=False):
    res = []
    labels = data[label_name]
    eig_values = None
    unlabeled_data = data.drop([label_name], axis=1)
    reduced_data = clf.fit_transform(unlabeled_data)
    score = 0
    if isPCA:
        score = clf.explained_variance_ratio_ 
        eig_values = clf.explained_variance_
    if isICA:
        score = clf.mixing_  
    return reduced_data, score, eig_values
```

```python
def get_recontrucation_error(transformer, data):
    transformer = transformer.fit(data)
    new_data = transformer.transform(data)
    inverse_components = np.linalg.pinv(transformer.components_)
    reconstructed_instances = utils.extmath.safe_sparse_dot(new_data, inverse_components.T)
    mse = ((data - reconstructed_instances) ** 2).mean()
    return mse
```

```python
def plot_bar(objects, performance, title=""):
    y_pos = np.arange(len(objects))
    plt.bar(y_pos, performance, align='center', alpha=0.5)
    plt.xticks(y_pos, objects, rotation='vertical')
    plt.title(title)
#     plt.ylim((-3,6))
    plt.show()
```

```python
for n in range(7,8):
    new_data = pd.DataFrame()
    for col in adults_x.columns:
        new_data[col] = adults_x[col] - np.mean(adults_x[col])
    pca = PCA(n_components=n)
    re_err = get_recontrucation_error(pca, new_data)
    objects = re_err.keys()
    performance = [re_err[key] for key in objects]
    plot_bar(objects, performance, title="MSE Reconstruction for Adults data for {} PCA components".format(n))
```

```python
def plot_mse(clf, data, title=""):
    re_err = get_recontrucation_error(clf, data)
    objects = re_err.keys()
    performance = [re_err[key] for key in objects]
    plot_bar(objects, performance, title=title)
```

```python
def plot_ica_weights(data, label_name):
    ica = FastICA(n_components=4, whiten=True)
    data_x = data.drop([label_name], axis=1)
    reduced_data, ica_comp, _, = appply_dim_reduction(data, ica, label_name, isICA=True)
    plot_mse(ica, data_x, title="")
    ica_comp = ica_comp.T
    for i in range(len(ica_comp)):
        plot_bar(data_x.columns, ica_comp[i], title="ICA component {} for WDBC".format(i+1))
plot_ica_weights(adults_data, 'income')
```

### Randomized Projections (RP)

```python
def RP(data, title=""):
    max_attemps=8
    for i in range(2,max_attemps):
        transformer = GaussianRandomProjection(eps=0.5, n_components=i)
        for j in range(100):
            transformer = transformer.fit(data)
        plot_mse(transformer, data, title=title+ " for " + str(i) + " components")
```

# Feature Selection using Variance Threshold

```python
def get_best_k(X, y, k=3):
    clf = SelectKBest(chi2, k=k).fit(X, y)
    return clf, clf.transform(X), clf.scores_
```

```python
clf, reduced_data, score = get_best_k(adults_x, adults_y, k=3)
plot_bar(adults_x.columns, score)
```
